```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(5)
```

```python
driver.get('https://jkhan.live/')
```

```python
menu = driver.find_element(By.XPATH, "//span[@class='duration-300 hover:-rotate-90 cursor-pointer m-2']").click()
```

```python
projects = driver.find_element(By.XPATH, "//*[contains(text(), 'Projects')]")
projects.click()
```

```python
project_1_test = driver.find_element(By.XPATH, "//h2[contains(text(), 'The Smartest City')]")
project_1_test.click()
```

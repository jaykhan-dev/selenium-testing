# First test case 

## Amazon

Task: Search for 'horse bow' on Amazon and verify that the search returns the input string in the results section. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(5)
```

```python
driver.get('https://www.amazon.com/')
```

```python
search = driver.find_element(By.ID, "twotabsearchtextbox")
search.send_keys('horse bow', Keys.ENTER)
```

```python
expected_text = '"horse bow"'
actual_text = driver.find_element(By.XPATH, "//span[@class='a-color-state a-text-bold']").text
```

```python
assert expected_text == actual_text, f'Error.  Expected text: {expected_text}, but actual text: {actual_text}'
```
